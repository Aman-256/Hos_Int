```python
import requests
import pandas as pd
from datetime import datetime, timedelta

# Replace with your Visual Crossing API key
API_KEY = 'FZ3C63VAEDD832YLWTEDRDKX5'

# Base URL for the Visual Crossing API
BASE_URL = 'https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline'

# Function to fetch historical weather data for a given date
def fetch_weather_data(location, date):
    url = f"{BASE_URL}/{location}/{date}/{date}?key={API_KEY}&include=obs"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        # Extract relevant fields: temperature, humidity, wind speed, and precipitation
        if 'days' in data and len(data['days']) > 0:
            weather = data['days'][0]
            return {
                'Date': date,
                'Temp': weather.get('temp', None),
                'Humidity': weather.get('humidity', None),
                'WindSpeed': weather.get('windspeed', None),
                'Rain': weather.get('precip', None),  # Precipitation (rainfall)
            }
    else:
        print(f"Failed to fetch data for {date}")
    return None

# Define the location (e.g., Dallas, TX) and date range
location = 'Dallas,TX'
start_date = datetime(2022, 1, 1)
end_date = datetime(2022, 12, 31)

# Fetch data for each day in the specified range
weather_data = []
current_date = start_date
while current_date <= end_date:
    formatted_date = current_date.strftime('%Y-%m-%d')
    weather_info = fetch_weather_data(location, formatted_date)
    if weather_info:
        weather_data.append(weather_info)
    current_date += timedelta(days=1)

# Convert the data into a DataFrame
weather_df = pd.DataFrame(weather_data)

# Save the data to a CSV for future use
weather_df.to_csv('weather_data.csv', index=False)

# Display the fetched data
print(weather_df.head())
```
